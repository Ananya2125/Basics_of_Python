```python
#2^n-complex
#2^n -1      no.of disc shifts
#tower of hanoi
def towerOfHanoi(n,s,d,a):
  if n==1:
    print("Move disc 1 from ",s," to ",d)
    return
  towerOfHanoi(n-1,s,a,d)
  print("Move disc ",n," from ",s," to ",d)
  towerOfHanoi(n-1,a,d,s)

n=int(input("Enter the number of disc: "))
towerOfHanoi(n,'A','C','B')
```

```python
#binary search
def binary(a,l,h,x):
  while(l<=h):
    mid=int((l+h)/2)
    if a[mid]>x:
      h=mid-1
    elif a[mid]<x:
      l=mid+1
    else:
      return mid
  return -1

a=[]
n=int(input("enter the number of elements in array: "))
for j in range(n):
  ele=int(input())
  a.append(ele)
x=int(input("Enter the number to be searched: "))
p=binary(a,0,n-1,x)
if p!=-1:
  print("Found at ",p+1)
else:
  print("not found")
  
```

```python
#selection sort
def selection(a,n):
  for i in range(n):
    min=i
    for j in range(i+1,n):
      if(a[j]<a[min]):
        min=j
    (a[i],a[min])=(a[min],a[i])

a=[]
n=int(input("enter the number of elements in array: "))
for j in range(n):
  ele=int(input())
  a.append(ele)

selection(a,n)
print(a)
```

```python
#mergesort
def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m
 
    # create temp arrays
    L = [0] * (n1)
    R = [0] * (n2)
 
    # Copy data to temp arrays L[] and R[]
    for i in range(0, n1):
        L[i] = arr[l + i]
 
    for j in range(0, n2):
        R[j] = arr[m + 1 + j]
 
    # Merge the temp arrays back into arr[l..r]
    i = 0     # Initial index of first subarray
    j = 0     # Initial index of second subarray
    k = l     # Initial index of merged subarray
 
    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
 
    # Copy the remaining elements of L[], if there
    # are any
    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
 
    # Copy the remaining elements of R[], if there
    # are any
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1
 
# l is for left index and r is right index of the
# sub-array of arr to be sorted
 
 
def mergeSort(arr, l, r):
    if l < r:
 
        # Same as (l+r)//2, but avoids overflow for
        # large l and h
        m = l+(r-l)//2
 
        # Sort first and second halves
        mergeSort(arr, l, m)
        mergeSort(arr, m+1, r)
        merge(arr, l, m, r)
 
arr=[]
n=int(input("enter the number of elements in array: "))
for p in range(n):
  ele=int(input())
  arr.append(ele)

mergeSort(arr,0,n-1)
for x in range(n):
  print(arr[x],end=" ")
```

```python
#bubblesort
def bubble(a):
  n=len(a)
  s=False
  for i in range(n-1):
    for j in range(n-i-1):
      if a[j]>a[j+1]:
        s=True
        a[j],a[j+1]=a[j+1],a[j]
    if not s:
      return
a=[]
n=int(input("enter the number of elements in array: "))
for p in range(n):
  ele=int(input())
  a.append(ele)

bubble(a)
print(a)
#insertion sort
```
