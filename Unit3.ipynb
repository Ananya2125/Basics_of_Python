```python
#arbitrary arguments-when we are not sure that how many arguments our function would require
#types---
# *args-used to pass variable number of arguments as TUPLE
# **kwargs-used to pass variable number of arguments as KEY-VALUE PAIR
```

```python
def fruits(*args):
  for fruit in args:
    print(fruit)
fruits('Orange','banana','apple','grapes')
```

```python
def fruits(**kwargs):
  for fruit in kwargs.values():
    print(fruit)
fruits(f1="orange",f2="banana",f3="litchi",f4="guava")
```

```python
#PERFECT NUMBER (6=1+2+3)
def perfect(n):
  sum=0
  for i in range(1,n):
    if(n%i==0):
      sum=sum+i
  if(sum==n):
    return True
  else:
    return False
for j in range(1,1000):
  if(perfect(j)):
    print(j)
```

```python
a=3                   #global variable
print("Global: ",a)
def fun():
  a=4                  #local variable
  print("Local variable: ",a)
  def fun1():
    a=7
    print("Local to fun 1 only: ",a)
  fun1()
fun()
```

```python
#global-declared outside of a function
gv="ABES_EC"
def welcome():
  print(gv)
welcome()
```

```python
#local-declared inside of a function
def welcome():
  lv="ABES"
  print(lv)
welcome()
```

```python
#enclosed--used in nested functions,whose local scope is undefined
def welcome():
  college="ABES-EC"
  def welcome_again():
    nonlocal college
    college="ABES_EC,Ghaziabad"
  welcome_again()
  print(college)
welcome()
```
